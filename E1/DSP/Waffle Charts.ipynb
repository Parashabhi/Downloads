## Waffle Charts

1. What is a Waffle Chart?

   * A Waffle Chart is a gripping visualization technique that is normally created to display progress towards goals. or
     It represents proportions or percentages using squares or rectangles.
                   
   * For example, imagine we have a waffle chart with 100 squares. Each square represents 1% of the total.

2. Setting Up the Environment

    * To create waffle charts in Python, we need to install and import the required libraries: pandas, matplotlib, and pywaffle.
    
    * Installation: pip install pandas matplotlib pywaffle
    
    * Importing modules: import pandas as pd, 
                         import matplotlib.pyplot as plt, 
                         from pywaffle import Waffle

```python
import pandas as pd
import matplotlib.pyplot as plt
from pywaffle import Waffle
```

3. Creating a Basic Waffle Chart

    * We can construct a basic waffle chart using random data to visualize proportions.
    * For example, let's say we want to represent the proportion of four categories: A, B, C and D with proportions 10%, 20%,  30% and 40% respectively.
    * We create a waffle chart with 100 squares, where 10 squares represent category A, 20 squares represent category B, 30 squares represent category C   and 40 squares represent category D.

```python
# Data
data = {'Categories': ['A', 'B', 'C','D'],
        'Proportions': [10,20,30,40]}

df = pd.DataFrame(data)

# Plotting the waffle chart
fig = plt.figure(
    FigureClass=Waffle,
    rows=10,
    values=df['Proportions'],
    labels=list(df['Categories'])
)

# Adding a title
plt.title('Our New Basic Waffle Chart')

# Displaying the chart
plt.show()
```

4. Customizing Waffle Charts

    * Waffle charts can be customized to suit specific needs and enhance visual appeal.
    * We can change the size and shape of waffle charts, modify colors, labels, and legends to provide more context.
    * For example, we can adjust the number of rows and columns, change the color palette, add labels to each square, and include a legend to explain the categories represented.

```python
# Data
data = {'Categories': ['A', 'B', 'C'],
        'Proportions': [40, 30, 30]}

df = pd.DataFrame(data)

# Plotting the waffle chart
fig = plt.figure(
    FigureClass=Waffle,
    rows=3,
    columns=6,  # Adjust the number of columns
    values=df['Proportions'],
    labels=list(df['Categories']),
    colors=['#FF5733', 'm', 'k'],  # Custom color palette
)

# Adding a title
plt.title("Customized Waffle Chart")

# Displaying the chart
plt.show()
```

5. Real-Life Examples

    * Example 1: Visualizing market share of smartphone brands.

        * We can create a waffle chart to represent the market share of different smartphone brands such as Xiaomi, Samsung, Apple, and Realme.
        * The number of squares allocated to each brand corresponds to its market share percentage.
    * Example 2: Representing gender distribution in a population.

        * Suppose we want to show the gender distribution in a population with proportions of 55% female and 45% male.
          Using a waffle chart, we can visually represent these proportions with squares allocated accordingly.
    * Example 3: Showing the composition of a product's customer base.

        * If we have information about a product's customer base, such as the percentage of customers in different age groups (e.g., 18-25, 26-35, 36-45, and so on), we can use a waffle chart to present this composition.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pywaffle import Waffle

# Data
data = {'Categories': ['Realme', 'Samsung', 'Vivo', 'Apple', 'JIo'],
        'Proportions': [10, 30, 15, 35, 10]}

df = pd.DataFrame(data)

# Plotting the waffle chart
fig = plt.figure(
    FigureClass=Waffle,
    rows=5,
    values=df['Proportions'],
    labels=list(df['Categories']),
    figsize=(5, 5),  # Optional: Adjust the figure size
    colors=['#FF5733', '#FFC300', '#C70039', '#900C3F', '#FF00FF']  # Optional: Custom colors
)

# Adding a title
plt.title("Waffle Chart")

# Displaying the chart
plt.show()
```
